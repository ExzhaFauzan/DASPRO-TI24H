```python
import numpy as np
import pandas as pd

jmlData = 1000
np.random.seed(42)

usia = np.random.randint(17, 61, size=jmlData)
gender = np.random.choice(['L', 'P'], size=jmlData)
durasi_dengar = np.random.randint(10, 301, size=jmlData)  
frekuensi_dengar = np.random.randint(1, 31, size=jmlData)  
rating = np.random.choice(['Buruk', 'Sedang', 'Bagus'], size=jmlData)

df = pd.DataFrame({
    'Usia': usia,
    'Gender': gender,
    'DurasiDengarPerHari': durasi_dengar,
    'FrekuensiDengarPerBulan': frekuensi_dengar,
    'RatingMusikAlternatif': rating
})

print(df)
print(df.info())  
print(df.head())  

df.to_csv("dataMusikAlternatif.csv", index=False)


print(f"Usia Pendengar Termuda: {df['Usia'].min()}")
print(f"Usia Pendengar Tertua: {df['Usia'].max()}")

laki = df[df['Gender'] == 'L']
print(f"Jumlah pendengar laki-laki: {laki.shape[0]}")
print(f"Rata-rata durasi dengar per hari (laki-laki): {laki['DurasiDengarPerHari'].mean():.2f} menit")
```
